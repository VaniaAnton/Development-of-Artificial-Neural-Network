```python
!pip install -q albumentations
import os, random, shutil, zipfile
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import albumentations as A
import cv2
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
# Download dataset (images and labels)
!wget -q https://www.robots.ox.ac.uk/~vgg/data/flowers/102/102flowers.tgz
!wget -q https://www.robots.ox.ac.uk/~vgg/data/flowers/102/imagelabels.mat
!tar -xvzf 102flowers.tgz

# Load labels
import scipy.io
labels = scipy.io.loadmat('imagelabels.mat')['labels'][0] - 1  # 0-indexed
image_paths = sorted([os.path.join('jpg', f) for f in os.listdir('jpg') if f.endswith('.jpg')])
data = list(zip(image_paths, labels))

# Filter top 10 classes with most images
from collections import Counter
class_counts = Counter(labels)
top_classes = [cls for cls, _ in class_counts.most_common(10)]
filtered_data = [(p, l) for p, l in data if l in top_classes]

# Relabel classes 0-9
label_map = {old: i for i, old in enumerate(top_classes)}
filtered_data = [(p, label_map[l]) for p, l in filtered_data]

# Split into train/val/test
random.shuffle(filtered_data)
train_data = filtered_data[:int(0.7*len(filtered_data))]
val_data   = filtered_data[int(0.7*len(filtered_data)):int(0.85*len(filtered_data))]
test_data  = filtered_data[int(0.85*len(filtered_data)):]
```

```python
IMG_SIZE = 128
BATCH_SIZE = 32
NUM_CLASSES = 10

# Albumentations augmentations
transform = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
    A.Rotate(limit=30, p=0.5),
    A.RandomCrop(height=IMG_SIZE, width=IMG_SIZE),
])

def preprocess(img_path, label):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    augmented = transform(image=img)['image']
    img = augmented / 255.0
    return img, tf.keras.utils.to_categorical(label, NUM_CLASSES)

def data_generator(data, batch_size):
    while True:
        random.shuffle(data)
        for i in range(0, len(data), batch_size):
            batch = data[i:i+batch_size]
            X, y = zip(*[preprocess(p, l) for p, l in batch])
            yield np.array(X), np.array(y)

train_gen = data_generator(train_data, BATCH_SIZE)
val_gen   = data_generator(val_data, BATCH_SIZE)
test_X, test_y = zip(*[preprocess(p, l) for p, l in test_data])
test_X = np.array(test_X)
test_y = np.array(test_y)
```

```python
def build_custom_model():
    model = models.Sequential([
        layers.Conv2D(32, (3,3), padding='same', activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
        layers.Conv2D(32, (3,3), activation='relu'),
        layers.MaxPooling2D(),
        layers.Dropout(0.25),

        layers.Conv2D(64, (3,3), padding='same', activation='relu'),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D(),
        layers.Dropout(0.25),

        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    return model

model = build_custom_model()
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
steps_per_epoch = len(train_data) // BATCH_SIZE
val_steps = len(val_data) // BATCH_SIZE

checkpoint = keras.callbacks.ModelCheckpoint('best_custom_cnn.h5', save_best_only=True, monitor='val_loss', mode='min')
history = model.fit(
    train_gen,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_gen,
    validation_steps=val_steps,
    epochs=15,
    callbacks=[checkpoint]
)
```

```python
model.load_weights('best_custom_cnn.h5')
preds = model.predict(test_X)
y_pred = np.argmax(preds, axis=1)
y_true = np.argmax(test_y, axis=1)

print(classification_report(y_true, y_pred))
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
for i in range(5):
    idx = random.randint(0, len(test_X) - 1)
    img = test_X[idx]
    plt.imshow(img)
    pred = np.argmax(model.predict(np.expand_dims(img, 0)))
    prob = np.max(model.predict(np.expand_dims(img, 0)))
    true = y_true[idx]
    plt.title(f"Predicted: {pred}, True: {true}, Prob: {prob:.2f}")
    plt.axis('off')
    plt.show()
```

```python
base = keras.applications.EfficientNetB0(include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3), weights='imagenet')
base.trainable = False

inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = base(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

model_tl = keras.Model(inputs, outputs)
model_tl.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_tl.summary()

# Train top
history_tl = model_tl.fit(train_gen, steps_per_epoch=steps_per_epoch, validation_data=val_gen,
                          validation_steps=val_steps, epochs=5)
```
